## Fetch stock data

### Subtask:
Retrieve historical stock data for Rolls-Royce Holdings (RR.L) and Lockheed Martin (LMT) for the specified date range using a financial data library.


**Reasoning**:
Fetch the historical stock data for the specified tickers and date range using yfinance.



```python
import yfinance as yf
import pandas as pd

tickers = ['RR.L', 'LMT', 'RTX', 'HON', 'SAF.PA', 'GD', 'NOC', 'BA.L', 'AM.PA', 'LDO.MI', 'BA']
start_date = '2022-07-01'
end_date = '2025-06-30'

stock_data = yf.download(tickers, start=start_date, end=end_date)
display(stock_data.head())
```

## Process data

### Subtask:
Prepare the fetched data for plotting, ensuring both stocks are in the same DataFrame and handling any potential missing values.


**Reasoning**:
Extract the 'Close' prices for both tickers, create a new DataFrame, handle missing values, and display the head of the new DataFrame.



```python
close_prices = stock_data['Close']
close_prices_df = pd.DataFrame(close_prices)
close_prices_df = close_prices_df.fillna(method='ffill')
display(close_prices_df.head())
```

## Visualize data

### Subtask:
Create a line chart to compare the adjusted closing prices of the two stocks over time.


**Reasoning**:
Create a line chart to visualize the adjusted closing prices of the two stocks over time, adding title, axis labels, and a legend as instructed.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
for ticker in close_prices_df.columns:
    plt.plot(close_prices_df.index, close_prices_df[ticker], label=ticker)

plt.title('Stock Price Comparison (July 2022 - June 2025)')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   Historical stock data for Rolls-Royce Holdings (RR.L) and Lockheed Martin (LMT) was successfully fetched for the period between July 1, 2022, and June 30, 2025.
*   The closing prices for both stocks were extracted into a single DataFrame, indexed by date.
*   Missing values in the closing price data were handled using the forward fill method.
*   A line chart was successfully generated to visually compare the closing price trends of RR.L and LMT over the specified period.

# Task
Create a Python script that generates a chart comparing the percentage increase in stock prices for Rolls-Royce Holdings, Lockheed Martin, RTX, Honeywell, Safran, General Dynamics, Northrop Grumman, BAE Systems, and Dassault Aviation, and Leonardo between July 2022 and June 2025.

## Calculate percentage change

### Subtask:
Calculate the percentage change in closing price for each ticker relative to the starting date.


**Reasoning**:
Calculate the percentage change for each ticker relative to the starting date and display the head of the resulting DataFrame.



```python
percentage_change_df = close_prices_df.apply(lambda x: (x / x.iloc[0] - 1) * 100)
display(percentage_change_df.head())
```

## Visualize percentage change

### Subtask:
Create a line chart to compare the percentage increase of all tickers over time.


**Reasoning**:
Create a line chart to visualize the percentage change of all tickers over time with appropriate labels and title.



```python
plt.figure(figsize=(14, 8))
for ticker in percentage_change_df.columns:
    plt.plot(percentage_change_df.index, percentage_change_df[ticker], label=ticker)

plt.title('Stock Price Percentage Change Comparison (July 2022 - June 2025)')
plt.xlabel('Date')
plt.ylabel('Percentage Change')
plt.legend()
plt.grid(True)
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The percentage change in stock prices for all analyzed companies from July 2022 to June 2025 was successfully calculated and visualized.
*   The generated chart displays the percentage change over time for each of the ten specified tickers, allowing for a direct visual comparison of their performance relative to the start date.

## Identify biggest earners (percentage-wise)

### Subtask:
Determine the companies with the largest percentage increase in stock price over the entire period.

**Reasoning**:
Extract the final percentage change for each ticker from the `percentage_change_df`, sort them in descending order, and display the result as a table.

```python
# Get the percentage change on the last day for each ticker
final_percentage_change = percentage_change_df.iloc[-1]

# Sort the tickers by percentage change in descending order
biggest_earners = final_percentage_change.sort_values(ascending=False)

# Create a mapping of tickers to company names
company_names = {
    'RR.L': 'Rolls-Royce Holdings',
    'LMT': 'Lockheed Martin',
    'RTX': 'RTX Corporation',
    'HON': 'Honeywell',
    'SAF.PA': 'Safran',
    'GD': 'General Dynamics',
    'NOC': 'Northrop Grumman',
    'BA.L': 'BAE Systems',
    'AM.PA': 'Dassault Aviation',
    'LDO.MI': 'Leonardo',
    'BA': 'Boeing',
    # Add other tickers and their company names if needed
}

# Create a DataFrame from the biggest earners Series
biggest_earners_df = biggest_earners.to_frame(name='Percentage Change')

# Add the "Company" column using the mapping
biggest_earners_df['Company'] = biggest_earners_df.index.map(company_names)

# Reorder columns to have "Company" first
biggest_earners_df = biggest_earners_df[['Company', 'Percentage Change']]

# Display the result as a table
print("Biggest Earners (Percentage Change):")
display(biggest_earners_df)
```

```python
# prompt: Using dataframe biggest_earners_df: Create a chart with vertical bars reflecting the gains in %

import altair as alt
# Create a bar chart of percentage change by ticker
chart = alt.Chart(biggest_earners_df.reset_index()).mark_bar().encode(
    x=alt.X('Ticker'), # Use Ticker for the x-axis
    y=alt.Y('Percentage Change') # Use Percentage Change for the y-axis
).properties(
    title='Percentage Change by Ticker' # Add a title to the chart
)
chart
```
